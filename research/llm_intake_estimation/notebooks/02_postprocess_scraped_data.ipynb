```python
import pandas as pd
```

```python
df_paths = [
    "../data/raw/Raw Ingredients.csv",
    "../data/raw/Minimally Processed Foods.csv",
    "../data/raw/Basic Processed Foods.csv",
    "../data/raw/Moderately Processed Foods.csv",
    "../data/raw/Highly Processed Foods.csv",
    "../data/raw/Specialty Foods.csv",
    "../data/raw/Beverages.csv",
    "../data/raw/Composite Dishes.csv"
]
df_names = [
    "Raw Ingredients",
    "Minimally Processed Foods",
    "Basic Processed Foods",
    "Moderately Processed Foods",
    "Highly Processed Foods",
    "Specialty Foods",
    "Beverages",
    "Composite Dishes"
]
```

```python
dfs = []
for path, name in zip(df_paths, df_names):
    df = pd.read_csv(path)
    df["category"] = name
    dfs.append(df)
df = pd.concat(dfs)
```

```python
df.columns
```

```python
# missing data in crucial columns
print("Calories: ", df["energy_kcal_100g"].isna().sum())
print("Carbohydrates: ", df["carbohydrates_100g"].isna().sum())
print("Fat: ", df["fat_100g"].isna().sum())
print("Proteins: ", df["proteins_100g"].isna().sum())
```

```python
is_missing_data = df["carbohydrates_100g"].isna() | df["fat_100g"].isna() | df["proteins_100g"].isna()
print(df[is_missing_data].shape)
df[is_missing_data]
```

```python
df = df[~is_missing_data].copy()
```
